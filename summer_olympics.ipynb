## **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
x=df[df["Medal"]=="Gold"]
```

```python
x
```

```python
var=[]
for sport in x["Sport"].unique():
  var.append([sport,len(x[x["Sport"]==sport])])
```

```python
var
```

```python
len(var)
```

```python
pd.DataFrame(var,columns=["Sport","Freq"]).sort_values(by="Freq",ascending=False).head()
```

```python
print("Aquatics is having most number of gold medals that is 1421.")
```

```python
b=pd.DataFrame(var,columns=["Sport","Freq"]).sort_values(by="Freq",ascending=False).head().plot(x="Sport",y="Freq",kind="bar",figsize=(7.5,5))
```
